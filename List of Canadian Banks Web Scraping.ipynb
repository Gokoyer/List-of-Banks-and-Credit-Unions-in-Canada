```python
#Using Beautiful soup for this table
#List of Big Five Banks in Canada
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_banks_and_credit_unions_in_Canada'
page= requests.get(url)
```

```python
soup = BeautifulSoup(page.text, "html")
```

```python
print(soup)
```

```python
soup.find_all('table')
```

```python
Big_5_table= soup.find_all('table')[0]
```

```python
Big_5_table
```

```python
Headings_Big_5 = Big_5_table.find_all('th')
```

```python
Headings_Big_5
```

```python
Bank_Headings_Big_5 = [title.text for title in Headings_Big_5]
print(Bank_Headings_Big_5)
```

```python
Bank_Headings_Big_5 = [title.text.strip() for title in Headings_Big_5]
print(Bank_Headings_Big_5)
```

```python
import pandas as pd
```

```python
pd.DataFrame(columns = Bank_Headings_Big_5)
```

```python
df = pd.DataFrame(columns = Bank_Headings_Big_5)
df
```

```python
Big_5_table.find_all('tr')
```

```python
Data_in_column = Big_5_table.find_all('tr')
```

```python
for row in Data_in_column:
    Data_in_row = row.find_all('td')
    individual_row_data = [data.text.strip() for data in Data_in_row]
    print(individual_row_data)
```

```python
#this is to remove empty [] in the first row
for row in Data_in_column[1:]:
    Data_in_row = row.find_all('td')
    individual_row_data = [data.text.strip() for data in  Data_in_row]
    print(individual_row_data)
```

```python
for row in Data_in_column[1:]:
    Data_in_row = row.find_all('td')
    individual_row_data = [data.text.strip() for data in Data_in_row]

    length = len(df)
    df.loc[length] = individual_row_data
```
